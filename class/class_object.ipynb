```python
### Instance variable and methods
'''
__init__()  -> this is constructor used to define the property of the object
'''
class Car: 
    def __init__(self, name, noofwindow):
        self.name=name                 #self.name is a instance variable similarly self.windows
        self.window=noofwindow

    def get_noof_window(self):
        return  self.window

bmw=Car("bmw",2)

audi=Car("audi",4)
print(bmw.get_noof_window())
print(audi.get_noof_window())

     
```

```python


class BankAccount: 
    def __init__ (self, owner, balance=0):
        self.owner=owner
        self.balance=balance
        #self.account_holder.append({"name":self.owner, "balance": self.balance})
    
    def deposit(self, amount):
        self.balance+=amount
        print(f"{amount} is deposited . New balance is {self.balance}")
    
    def withdraw(self, amount):
        if amount> self.balance:
            print("Insufficient fund")
        else:
            self.balance-=amount
            print(f"{amount} is withdrawn , New balance is {self.balance}")
    def getBalance(self):
        return self.balance

p1=BankAccount("piyush", 1000)
print(p1.getBalance())

p1.deposit(100)
print(p1.getBalance())

p1.withdraw(500)
print(p1.getBalance())

```
