### Tuples

1. Tuples are ordered collection of items that are immutable. They are similar to the lists, but their immutablity makes them different
2. It is used when immutabliity and ordered collection of items is required. They are commonly used in data structures, function argument and return values and as dictionary keys.

```python
t=()
print(t)
print(type(t))
```

```python
l=list()
print(type(l))
l.extend([1,2,3])
print(l)
```

```python
t=tuple()
print(type(t))
```

```python
numbers=tuple([1,2,3,4])
numbers #(1,2,3,4)
k=list(numbers)
k
```

```python
## accessing tuple element

for x in numbers:
    print(x)

print(numbers[-1:])
print(numbers[2:5])
```

```python
## Tuple operation

mixed_tuple=(1,2,3, "piyush", "dubey")

print(numbers + mixed_tuple)

```

```python
## immuatable nature of tuple -> element can not be changed once it is assigned

numbers[1]="krish"

```

```python
## tuple method

print(numbers.count(1))

print(numbers.index(3)) ## give me the first index of the elements
```

```python
## packing and unpacking tuple

#packing 
packeed_tuple=1,"hello",3.14
print(packeed_tuple)


#unpacking
a,b,c=packeed_tuple
print(a)
print(b)
print(c)

#unpacking with *
numbers=(1,2,3,4,5)
first,*middle, last=numbers
print(first)
print(middle) #[2,3,4]
print(last)


#list unpacking
l=[1,2,3,4]
a,b,c,d=l
print(f"hii there",a)
```

```python
## Nested tuple 

lst=[[1,2,3,4],(8,9,10)]
print(lst[1][:1])


nested_tuple=((1,2,3,4),("a","b","c"))

print(nested_tuple[0])
print(nested_tuple[1][2])

for x in nested_tuple:
    for y in x:
        print(y, end=" ")
    print()
```
