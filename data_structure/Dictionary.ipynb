###  Dictionary
1. It is unodered collection of items. They store data in key-value pairs. Keys must be unique and immutable(eg- string, numbers, or tuples) while values can be of any type.

```python
#creating the dictionaries

d={}
print(d)

d1=dict()
print(d1)
```

```python
student={"name":"piyush", "age":12, "class":"12th"}
print(student)


print(student.get("name"))
```

```python
#Error
student={"name":"piyush", "age":12, "name":"12th"}
print(student)
```

```python
#accessing the dict element

student={"name":"piyush", "age":12, "class":"12th", "grade":"A"}

print(student.get("grade"))
print(student["age"])

# if there is no element present in the dict you can also provide the default value
print(student.get("last_name", "Not Available"))
```

```python
#Modifying the dict elements
#dict are mutable, so you can add , update and delete elements

student["age"]=33
student["address"]="india"

print(student)



del student["address"] # delete a key, value pair  in the dict

g=student.pop("age")  #pop and return the key from the dict
print(g)

print(student)
```

```python
#dict methods

print(student.keys())  #get all the keys 
print(student.values()) # get all the values

items=student.items() ## get all key value pairs


print(items)
print(type(items))

```

```python
## shallow copy

student

student_copy=student
print(student)
print(student_copy)

student_copy["name"]="ayush"
print(student)
print(student_copy)


student_copy=student.copy() ## shallow copy
print(student)
print(student_copy)
student_copy["name"]="krish"
print(student)
print(student_copy)

```

```python
## iterating over dict
##  use loop to iterate 

## iterating over keys

for keys in student.keys():
    print(keys)

# iterate over values
for value in student.values():
    print(value)

# iterate over key, value pair

for key, value in student.items():
    print(f"key : {key} , value: {value}")
```

```python
## Nested dictionary

students={
    "1":{
        "name":"piytush", "age":2
    },
    "2":{
        "name":"peter", "age":33
    }

}

print(students)
print(students.keys())

for key, value in students.items():
    print(f"key: {key} value:{value}")
    # print(students.get(key))

    # print(students.get(key)["name"])
    for k, v in value.items():
        print(f"{k}:{v}")
```

```python
## dict comphrehension

squares={x:x**2   for x in range(5)}
print(squares)
```

```python
## conditional dict comprehenhsion

evensquares={x:x**2   for x in range(5) if x%2==0}
print(evensquares)
```
