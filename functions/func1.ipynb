```python
def greeting(name="Guest"):
    return f"hello {name}"

greeting() #hello Guest
greeting("ayush")
```

```python
## variable lenght arguments
##Positional and keywords arguments


#position argument
def print_number(*args):
    print(type(args))
    for n in args:
        print(n)

print_number(1,2,3,4,5,6,"piyush")  #postional argument
```

```python
#keyword argument -- all the parameter will be in the form of key , value pair
def print_details(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}:{value}")

print_details(name="krish", age="32",country="india")
```

```python
def print_details(*args,**kwargs):

    for val in args:
        print(f"{val}")

    for key,value in kwargs.items():
        print(f"{key}:{value}")

print_details(1,2,3,4,5,"krish", name="krish", age="32",country="india")
```
