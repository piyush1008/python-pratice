```python
## variable lenght arguments
##Positional and keywords arguments


#position argument
def print_number(*args):
    print(type(args))
    for n in args:
        print(n)

print_number(1,2,3,4,5,6,"piyush")  #postional argument
```

```python
#keyword argument -- all the parameter will be in the form of key , value pair
def print_details(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}:{value}")

print_details(name="krish", age="32",country="india")
```

```python
def print_details(*args,**kwargs):

    for val in args:
        print(f"{val}")

    for key,value in kwargs.items():
        print(f"{key}:{value}")

print_details(1,2,3,4,5,"krish", name="krish", age="32",country="india")
```

### LAMBDA Function

They are small anonymous function defined using the lambda keyword. They can have any nuhmber of arguments but only one expression. They are commmonly used for short operations or as argumetns to higher-order functions.



#Syntacx lambda arguments: expression

```python
def addition(a,b):
    return a+b

addition(2,3)

addition1=lambda a,b: a+b
print(addition(5,6))
```

```python
def even(num):
    if num%2==0:
        return True

even1=lambda num: num%2==0

print(even1(11))




```

```python
numbers=[1,2,3,4]

list(map(lambda x: x**2, numbers))


```

### Map function
It applies a given function to all items in an input list (or any other iterable) and returns a map object(an iterable). This is particualarly useful for transforming data in a list comprehensively.

```python
def square(x):
    return x*x
square(10)

numbers=[1,2,3,4]

l=list(map(square,numbers))


print(l)

## same things with lambda function
sq=lambda x : x**2

k=list(map(sq, numbers))
print(k)
```

```python
num1=[1,2,3]
num2=[4,5,6]

added_number=list(map(lambda x, y: x+y, num1,num2 ))
print(added_number)
```

```python
words=["apple", "banana", "cherry"]
upper_word=list(map(str.upper, words))

print(upper_word)
```

```python
def get_name(person):
    return person["name"]

people=[
    {"name":"piyush", "age":34},
    {"name":"ayush", "age":24}
]


l=list(map(get_name, people))
print(l)
```
