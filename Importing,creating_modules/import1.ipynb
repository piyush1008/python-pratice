```python
## Data serialization
#converting the data into the json

import json
data={"name":"piyush", "age":25}

json_str=json.dumps(data)  #converting the object into json string
print(json_str)
print(type(json_str))


c=json.loads(json_str)  #converting json string back to dict
print(type(c))
print(c)
```

```python
## csv
import csv 
with open("example.csv", "w", newline="") as file:
    writer=csv.writer(file)
    writer.writerow(['name', 'age'])
    writer.writerow(['piyuhs',34])

with open("example.csv","r") as file:
    reader=csv.reader(file)
    for row in reader:
        print(row)
```
