```python
# read whole file

with open("example.txt", "r") as file:
    content=file.read()
    print(content)
    
```

```python
# read line by line from a file
with open("example.txt", "r") as file:
    for content in file:
        print(content.strip())  ## strip() removed the new line character
```

```python
#writng the file (overwrite)
with open("example.txt", "w") as file:
    file.write("hello world\n")
    file.write("this is again a new line\n")
```

```python
#writign the file in the append mode(without overwrite)
with open("example.txt", "a") as file:
    file.write("appending the file\n")
```

```python
#writing multiple lines
lines=["firs line\n", "second line\n", "hii there piyush\n"]

with open("example.txt", "a") as file:
    file.writelines(lines)
```

```python
## Binary files 

data= b'\x00\x01\x02\x03\x04'
with open("example.bin", "wb") as file:
    file.write(data)
```

```python

def count_text_file(file_path):
    with open(file_path, "r") as file:
        lines=file.readlines()   # this will return an array ["this is first line\n", "this is second line\n"]
        line_count=len(lines)
        word_count=sum(len(line.split()) for line in lines)
        char_count=sum(len(line) for line in lines)
    return line_count, word_count, char_count
file_path="example.txt"

lines, words, characters=count_text_file(file_path)
print(f"lines: {lines} words: {words} , characters: {characters}")
```

```python
### writing and then reading a file 
#w+ mode in python is used to open a file for both reading and writing . If the file does not exist , it will be created. if the file 
# exists , its content will be truncated(i.e the file will be overwritten)
with open("example.txt", "w+") as file:
    file.write("hello world\n")
    file.write("this is a new line\n")

    ##move the file cursor to the beginning 
    file.seek(0)

    ## read the content of the file
    content=file.read()
    print(content)
```

```python
#checking if a path is a file or directory


import os

path="example.txt"

if os.path.isfile(path):
    print("This path is a file")
elif os.path.isdir(path):
    print("the path is directory")
else: 
    print("the path does not exist")
```
